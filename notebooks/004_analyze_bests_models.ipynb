```python
import json
import pandas as pd
```

```python
results_all = json.load(open('best_model/results/final_results.json', 'r'))
```

```python
res_raw = []
for r in results_all:
    best_model = 0
    current_score = 0
    for ii in range(1, len(r)):
        if current_score < r[ii]['best_score_']:
            best_model = ii
            current_score = r[ii]['best_score_']
    best_model = r[best_model]
    filter_info = r[0]['candidate']
    res_raw.append({
        'model': best_model['name'],
        'best_params': best_model['best_params_'],
        'best_score': best_model['best_score_'],
        'low_cutoff': filter_info['low_cutoff'],
        'high_cutoff': filter_info['high_cutoff'],
        'tmin': filter_info['tmin'],
        'tmax': filter_info['tmax'],
        'ica': filter_info['ica'],
        'notch': filter_info['notch'],
    })
```

```python
df_result = pd.DataFrame(res_raw, columns=['model', 'best_params', 'best_score', 'low_cutoff', 'high_cutoff', 'tmin', 'tmax', 'ica', 'notch'])
df_result
```
